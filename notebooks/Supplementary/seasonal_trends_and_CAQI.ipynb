## Seasonal trends (NO2 and O3)

Different pollutants experience seasonal variance, according to metereological conditions. Here we use NO2 and O3 to check its presence and assess its magnitude (the same plots for PM10 and PM2.5 are avaiable at `results/plots/seasonal_trends`).

### NO2
Below we plot the average day average value for each season in each city:

```python
display(Image(filename="../../results/plots/seasonal_trends/seasonal_trends_no2.png", width=1000))
```

The seasonal effect is present, and varies between cities. Cities with similar climate shows more similar variance (for example: Turin and Milan i the north, with colder winters have higher concentration and variance, Palermo and Naples in the south, which are warmer, experience lower concentration and variance, while Rome and Florence in the center have intermediate conditions). Similar patterns emerge for PM10 and PM2.5 (see `results/plots/seasonal_trends`)

### O3
Ozone, as opposed to NO2, tipically rise during summer, as we can see from the figure below:

```python
display(Image(filename="../../results/plots/seasonal_trends/seasonal_trends_o3.png", width=1000))
```

Here again greater variance is present in notherner cities, where winter is much colder.

For more details about how we calculated CAQI see `docs/methodology.md`
